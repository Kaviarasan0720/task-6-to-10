# task 4,question 15


Write a Pandas program to get the numeric representation of an array by identifying distinct values
of a given column of a dataframe. Sample Output:
Original DataFrame:
Name Date_Of_Birth Age
0 Alberto Franco 17/05/2002 18.5
1 Gino Mcneill 16/02/1999 21.2
2 Ryan Parkes 25/09/1998 22.5
3 Eesha Hinton 11/05/2002 22.0
4 Gino Mcneill 15/09/1997 23.0
Numeric representation of an array by identifying distinct values:
[0 1 2 3 1]
Index(['Alberto Franco', 'Gino Mcneill',



```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill'],
        'Date_Of_Birth': ['17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
        'Age': [18.5, 21.2, 22.5, 22.0, 23.0]}

df = pd.DataFrame(data)

# Get the numeric representation of an array by identifying distinct values
numeric_representation = df['Name'].astype('category').cat.codes

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# Print the numeric representation of an array
print("\nNumeric representation of an array by identifying distinct values:")
print(numeric_representation)

# Print the distinct values corresponding to the numeric representation
print("Index:", df['Name'].astype('category').cat.categories)
```

a. Write a NumPy program to convert an array to a float type

```python
import numpy as np

# Create an empty array
empty_array = np.empty((3, 3))
print("Empty Array:")
print(empty_array)

# Create a full array with a specific value
full_array = np.full((2, 4), 7.5)
print("\nFull Array:")
print(full_array)
```

c. Write a NumPy program to convert a list and tuple into arrays

```python
import numpy as np

# Convert a list to array
list_data = [1, 2, 3, 4, 5]
array_from_list = np.array(list_data)

# Convert a tuple to array
tuple_data = (1.5, 2.5, 3.5, 4.5, 5.5)
array_from_tuple = np.array(tuple_data)

# Print the results
print("Array from List:")
print(array_from_list)

print("\nArray from Tuple:")
print(array_from_tuple)
```

d. Write a NumPy program to find the real and imaginary parts of an array of complex numbers

```python
import numpy as np

# Create an array of complex numbers
complex_array = np.array([1 + 2j, 3 - 4j, 5 + 6j])

# Extract real and imaginary parts
real_parts = np.real(complex_array)
imaginary_parts = np.imag(complex_array)

# Print the results
print("Complex Array:")
print(complex_array)

print("\nReal Parts:")
print(real_parts)

print("\nImaginary Parts:")
print(imaginary_parts)
```

# task 4,question 8

8. Use the diabetes data set from UCI data set for performing the following:

Apply Univariate analysis:  Frequency
 Mean,  Median,  Mode,  Variance  Standard Deviation
 Skewness and Kurtosis

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Load the diabetes dataset
# Replace 'diabetes.csv' with the actual file path or URL
df = pd.read_csv('/content/diabetes.csv')

# Display the first few rows of the dataset
print("Sample of the Diabetes Dataset:")
print(df.head())

# Univariate Analysis
# Frequency
print("\nFrequency:")
print(df['Age'].value_counts())

# Mean
print("\nMean:")
print(df['Age'].mean())

# Median
print("\nMedian:")
print(df['Age'].median())

# Mode
print("\nMode:")
print(df['Age'].mode())

# Variance
print("\nVariance:")
print(df['Age'].var())

# Standard Deviation
print("\nStandard Deviation:")
print(df['Age'].std())

# Skewness
print("\nSkewness:")
print(skew(df['Age']))

# Kurtosis
print("\nKurtosis:")
print(kurtosis(df['Age']))
```

 7. i. Using Pandas, Create a DataFrame with a list of dictionaries, row indices, and column
indices.

```python
import pandas as pd

# List of dictionaries
data = [
    {'Name': 'John', 'Age': 25, 'City': 'New York'},
    {'Name': 'Alice', 'Age': 30, 'City': 'San Francisco'},
    {'Name': 'Bob', 'Age': 22, 'City': 'Chicago'}
]

# Row indices
row_indices = ['Person1', 'Person2', 'Person3']

# Column indices
column_indices = ['Name', 'Age', 'City']

# Create a DataFrame
df = pd.DataFrame(data, index=row_indices, columns=column_indices)

# Display the DataFrame
print(df)
```

ii. Use index label to delete or drop rows from a Pandas DataFrame. 8. Apply and explore various plotting functions on UCI data sets

```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 30, 22],
        'City': ['New York', 'San Francisco', 'Chicago']}

df = pd.DataFrame(data, index=['Person1', 'Person2', 'Person3'])

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Drop a row by index label
df = df.drop('Person2')

# Display the DataFrame after dropping a row
print("\nDataFrame after dropping 'Person2':")
print(df)
```
